```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('wfk')
```

```python
xgboost_error_path = "../Outputs/Fitted Models/XGBoost_HalvingGridSearchCV_20250313_163658/error_df.csv"
df_e_xg = pd.read_csv(xgboost_error_path)
enet_error_path = "../Outputs/Fitted Models/ElasticNet_HalvingGridSearchCV_20250313_162809/error_df.csv"
df_e_enet = pd.read_csv(enet_error_path)

display(df_e_xg.head())
```

```python
fig, axs = plt.subplots(1,figsize=(8,3))

width = 0.3

x = np.arange(len(df_e_xg['Error Category']))

# plot xgboost data
axs.bar(x - width/2, df_e_xg['RMSE [kcal/mol]'], width, label='XGBoost')
# plot elastic net data
axs.bar(x + width/2, df_e_enet['RMSE [kcal/mol]'], width, label='ElasticNet')

axs.set_xticks(x)
axs.set_xticklabels(df_e_xg['Error Category'], rotation=-90, fontsize=12)
axs.set_ylabel('Prediction RMSE [kcal/mol]', fontsize=12)
axs.legend(frameon=True, fontsize=10)
```
